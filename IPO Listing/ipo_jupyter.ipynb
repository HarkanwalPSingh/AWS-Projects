```python
import requests
import bs4
import pandas as pd
from datetime import date
```

```python
res = requests.get('https://investorzone.in/ipo/')
soup = bs4.BeautifulSoup(res.text,"lxml")
```

```python
items_mainline = soup.find(id='table-fill-1')
items_sme = items = soup.find(id='hot-stocks')
```

```python
headers = items_mainline.find_all('th')
```

```python
headers
```

```python
ipo_map = {}
```

```python
for x in headers:
    ipo_map[x.text] = []
```

```python
ipo_map
```

```python
rows = items_mainline.find_all('tr')
```

```python
rows.extend(items_sme.find_all('tr'))
```

```python
len(rows)
```

```python
rows
```

```python
row_list = []
for row in rows:
    row_list.append(row.findAll('td'))

row_list = list(filter(lambda x: x != [], row_list))
```

```python
for x in row_list:
    for row_element in x:
        print(row_element.text)
```

```python
def get_null_index(row_list):
    null_index = []
    for i,x in enumerate(row_list):
        if x == []:
            null_index.append(i)
    
    return null_index
```

```python
null_index = get_null_index(row_list)
print(null_index)
```

```python
for i in range(len(row_list)):
    for j,x in enumerate(headers):
        temp = ipo_map[x.text]
        temp.append(row_list[i][j].text)
        ipo_map[x.text] = temp
```

```python
ipo_map
```

```python
df = pd.DataFrame(ipo_map)
```

```python
df
```

```python
today = date.today()
today = today.strftime("%Y-%m-%d")
curr_date = pd.to_datetime(today)
```

```python
df["Open Date"] = pd.to_datetime(df["Open Date"])
df["Close Date"] = pd.to_datetime(df["Close Date"])
```

```python
# upcoming_ipos = df[(df["Open Date"] <= curr_date) & (curr_date <= df["Close Date"])]
upcoming_ipos = df[curr_date <= df["Close Date"]]
```
